```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf

# 设置随机种子，保证结果可重复
np.random.seed(42)

print("库导入成功！")
```

```python
# 设置参数
beta_0 = 1      # 截距
beta_1 = 2      # 斜率
n = 100         # 样本量
sigma = 1       # 误差标准差

# 生成 x 变量（均匀分布）
x = np.random.uniform(0, 10, n)

# 生成误差项 epsilon ~ N(0, 1)
epsilon = np.random.normal(0, sigma, n)

# 生成 y
y = beta_0 + beta_1 * x + epsilon

# 查看前几行数据
print("前5个样本:")
for i in range(5):
    print(f"x[{i}]={x[i]:.4f}, y[{i}]={y[i]:.4f}")

# 可视化
plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.6, label='模拟数据')
plt.plot(x, beta_0 + beta_1 * x, color='red', label='真实回归线')
plt.xlabel('x')
plt.ylabel('y')
plt.title('模拟的线性回归数据')
plt.legend()
plt.show()
```
